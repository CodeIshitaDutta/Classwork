```python
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import numpy as np

Kine_data = pd.read_csv('Kinematics_Data.csv')
```

```python
Kine_data.head()
```

```python
Kine_data['datetime'] = Kine_data['date'] + ' ' + Kine_data['time']
```

```python
Kine_data['datetime'] = pd.to_datetime(Kine_data['datetime'], format = '%Y-%m-%d %H:%M:%S:%f')
```

```python
Kine_data = Kine_data.sort_values(by = 'datetime')
```

```python
Kine_data['int_datetime'] = Kine_data['datetime'].apply(lambda x: x.value)
```

```python
Kine_data['gyro_mag'] = Kine_data['gyro_x']**2 + Kine_data['gyro_y']**2 + Kine_data['gyro_z']**2
Kine_data['gyro_mag'] = Kine_data['gyro_mag']**(1/2)
```

```python
Kine_data.head()
```

```python
Kine_data.info()
```

```python
def plot_each_day(data,day_col,time_col,y_col,color='none'):
    days = data[day_col].unique()
    for day in days:
        day_data = data[data[day_col] == day]
        x = day_data[time_col]
        y = day_data[y_col]
        if color == 'none':
            plt.scatter(x,y)
        else:
            colors = {0:'yellow',1:'green'}
            plt.scatter(x,y,c=day_data[color].map(colors))
        plt.title(day)
        plt.show()
        print(day_data[time_col].max() - day_data[time_col].min())
    return
```

```python
walk_data = Kine_data[Kine_data['activity'] == 0]
leftwrist_walk = walk_data[walk_data['wrist'] == 0]
rightwrist_walk = walk_data[walk_data['wrist'] == 1]
run_data = Kine_data[Kine_data['activity'] == 1]
leftwrist_run = run_data[run_data['wrist'] == 0]
rightwrist_run = run_data[run_data['wrist'] == 1]
```

```python
walk_data['date'].unique()
```

```python
run_data['date'].unique()
```

```python
plot_each_day(walk_data,'date','datetime','gyro_z')
```

```python
plot_each_day(run_data,'date','datetime','gyro_z')
```

```python
plot_each_day(Kine_data,'date','datetime','gyro_z',color = 'activity')
```

```python
test_data = Kine_data[Kine_data['date'] == '2017-7-4'][:321]
test_data.head()
```

```python
x_test = test_data['datetime']
y_test = test_data['gyro_z']
to_cluster = test_data['int_datetime']
plt.plot(x_test,y_test)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
to_cluster = np.array(to_cluster)
thirty_sec = pd.to_datetime('12:00:30').value - pd.to_datetime('12:00:00').value
print(thirty_sec)
HC_test=AgglomerativeClustering(n_clusters=None, distance_threshold=thirty_sec)

HC_test.fit(to_cluster.reshape(-1, 1))

plt.scatter(x_test, y_test, c=HC_test.labels_, cmap='rainbow')
plt.title('Test')
plt.show()
```
